```python
import gc
import glob
import geopandas as gpd


# SHAPE 파일 경로 리스트
file_path = glob.glob("./data/NGII_CDM_도로분리대.zip")[0]

# Shape 파일을 GeoDataFrame으로 불러오기(글씨 깨지면 인코딩 cp949로)
gdf = gpd.GeoDataFrame.from_file(file_path, encoding='utf-8')

# 경위도 좌표계 설정(좌표계 써져있으면 이용, 압축폴더 안에 prj파일 있으면 따로 설정 안해도 됨)
# gdf = gdf.to_crs({"init": "epsg:5179"})

gdf
```
